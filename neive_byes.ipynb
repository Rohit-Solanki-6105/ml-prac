## 1. Using the dataset provided below, use the Naive Bayes algorithm to predict the Class (mammal or non-mammal) for a new creature with the following attributes: 
- Give Birth: yes 
- Can Fly: no 
- Live in Water: yes 
- Have Legs: no 

```python
import pandas as pd
```

```python
data = [
    ["human","yes","no","no","yes","mammals"],
    ["python","no","no","no","no","non-mammals"],
    ["salmon","no","no","yes","no","non-mammals"],
    ["whale","yes","no","yes","no","mammals"],
    ["frog","no","no","sometimes","yes","non-mammals"],
    ["komodo","no","no","no","yes","non-mammals"],
    ["bat","yes","yes","no","yes","mammals"],
    ["pigeon","no","yes","no","yes","non-mammals"],
    ["cat","yes","no","no","yes","mammals"],
    ["leopard shark","yes","no","yes","no","non-mammals"],
    ["turtle","no","no","sometimes","yes","non-mammals"],
    ["penguin","no","no","sometimes","yes","non-mammals"],
    ["porcupine","yes","no","no","yes","mammals"],
    ["eel","no","no","yes","no","non-mammals"],
    ["salamander","no","no","sometimes","yes","non-mammals"],
    ["gila monster","no","no","no","yes","non-mammals"],
    ["platypus","no","no","no","yes","mammals"],
    ["owl","no","yes","no","yes","non-mammals"],
    ["dolphin","yes","no","yes","no","mammals"],
    ["eagle","no","yes","no","yes","non-mammals"]
]

# Create DataFrame
df = pd.DataFrame(data, columns=[
    "Name","Give Birth","Can Fly","Live in Water","Have Legs","Class"
])
```

```python
df
```

```python
# classes
class_counts = df["Class"].value_counts()
class_counts
```

```python
# total samples
total = len(data)
```

```python
total
```

```python
# prior probabilities
priors = class_counts / total
priors
```

```python
def likelihood(feature, value, class_label):
    subset = df[df["Class"] == class_label]
    count = len(subset[subset[feature] == value])
    total_class = len(subset)
    
    # laplace smoothing
    return (count + 1) / (total_class + len(df[feature].unique()))
```

```python
def predict(sample):
    results = {}

    for class_label in df["Class"].unique():
        prob = priors[class_label]

        prob *= likelihood("Give Birth", sample["Give Birth"], class_label)
        prob *= likelihood("Can Fly", sample["Can Fly"], class_label)
        prob *= likelihood("Live in Water", sample["Live in Water"], class_label)
        prob *= likelihood("Have Legs", sample["Have Legs"], class_label)

        results[class_label] = prob

    return results
```

```python
# new creature
new_creature = {
    "Give Birth": "yes",
    "Can Fly": "no",
    "Live in Water": "yes",
    "Have Legs": "no"
}

# new probabilities
result = predict(new_creature)
result
```

```python
# final prediction
final_class = max(result, key=result.get)
final_class
```

## 2. Using the Naive Bayes algorithm and the dataset below, predict whether the animal (black, ragged, small) is well-behaved or dangerous.

```python
data = [
    ["brown","ragged","small","well-behaved"],
    ["black","ragged","big","dangerous"],
    ["black","smooth","big","dangerous"],
    ["black","curly","small","well-behaved"],
    ["white","curly","small","well-behaved"],
    ["white","smooth","small","dangerous"],
    ["red","ragged","big","well-behaved"]
]

# Create DataFrame
df = pd.DataFrame(data, columns=["Color","Fur","Size","Class"])
df
```

```python
class_counts = df["Class"].value_counts()
total = len(data)

priors = class_counts / total
priors
```

```python
def predict_behave(sample):
    results = {}

    for class_label in df["Class"].unique():
        prob = priors[class_label]

        prob *= likelihood("Color", sample["Color"], class_label)
        prob *= likelihood("Fur", sample["Fur"], class_label)
        prob *= likelihood("Size", sample["Size"], class_label)

        results[class_label] = prob

    return results
```

```python
animal = {
    "Color": "black",
    "Fur": "ragged",
    "Size": "small"
}

result = predict_behave(animal)
result
```

```python
final_class = max(result, key=result.get)
final_class
```
