```python
import numpy as np
import pandas as pd
data = [
    ["human","yes","no","no","yes","mammals"],
    ["python","no","no","no","no","non-mammals"],
    ["salmon","no","no","yes","no","non-mammals"],
    ["whale","yes","no","yes","no","mammals"],
    ["frog","no","no","sometimes","yes","non-mammals"],
    ["komodo","no","no","no","yes","non-mammals"],
    ["bat","yes","yes","no","yes","mammals"],
    ["pigeon","no","yes","no","yes","non-mammals"],
    ["cat","yes","no","no","yes","mammals"],
    ["leopard shark","yes","no","yes","no","non-mammals"],
    ["turtle","no","no","sometimes","yes","non-mammals"],
    ["penguin","no","no","sometimes","yes","non-mammals"],
    ["porcupine","yes","no","no","yes","mammals"],
    ["eel","no","no","yes","no","non-mammals"],
    ["salamander","no","no","sometimes","yes","non-mammals"],
    ["gila monster","no","no","no","yes","non-mammals"],
    ["platypus","no","no","no","yes","mammals"],
    ["owl","no","yes","no","yes","non-mammals"],
    ["dolphin","yes","no","yes","no","mammals"],
    ["eagle","no","yes","no","yes","non-mammals"]
]

# Create DataFrame
df = pd.DataFrame(data, columns=[
    "Name","Give Birth","Can Fly","Live in Water","Have Legs","Class"
])
```

```python
# drop Name (not useful)
df = df.drop(columns=["Name"])
```

### encode categorical values -> integers

```python
for col in df.columns:
    df[col] = pd.factorize(df[col])[0]
```

```python
X = df.drop(columns=["Class"]).values
y = df["Class"].values
```

```python
# classes
class_counts = df["Class"].value_counts()
class_counts
```

```python
#priors
P_mammals = len(df[df["Class"]=="mammals"]) / len(df)
P_non = len(df[df["Class"]=="non-mammals"]) / len(df)
```

```python
mammals = df[df["Class"] == "mammals"]
non_mammals = df[df["Class"] == "non-mammals"]

# mammals
P_givebirth_m = len(mammals[mammals["Give Birth"] == "yes"]) / len(mammals)
P_fly_m       = len(mammals[mammals["Can Fly"] == "no"]) / len(mammals)
P_water_m     = len(mammals[mammals["Live in Water"] == "yes"]) / len(mammals)
P_legs_m      = len(mammals[mammals["Have Legs"] == "yes"]) / len(mammals)
```

```python
# non mammals
P_givebirth_n = len(non_mammals[non_mammals["Give Birth"] == "yes"]) / len(non_mammals)
P_fly_n       = len(non_mammals[non_mammals["Can Fly"] == "no"]) / len(non_mammals)
P_water_n     = len(non_mammals[non_mammals["Live in Water"] == "yes"]) / len(non_mammals)
P_legs_n      = len(non_mammals[non_mammals["Have Legs"] == "yes"]) / len(non_mammals)
```

```python
# new creature
new_creature = {
    "Give Birth": "yes",
    "Can Fly": "no",
    "Live in Water": "yes",
    "Have Legs": "no"
}
```

```python
mammals_score = (
    P_mammals *
    len(df[(df["Give Birth"]=="yes") & (df["Class"]=="mammals")]) / len(df[df["Class"]=="mammals"]) *
    len(df[(df["Can Fly"]=="no") & (df["Class"]=="mammals")]) / len(df[df["Class"]=="mammals"]) *
    len(df[(df["Live in Water"]=="yes") & (df["Class"]=="mammals")]) / len(df[df["Class"]=="mammals"]) *
    len(df[(df["Have Legs"]=="no") & (df["Class"]=="mammals")]) / len(df[df["Class"]=="mammals"])
)

non_score = (
    P_non *
    len(df[(df["Give Birth"]=="yes") & (df["Class"]=="non-mammals")]) / len(df[df["Class"]=="non-mammals"]) *
    len(df[(df["Can Fly"]=="no") & (df["Class"]=="non-mammals")]) / len(df[df["Class"]=="non-mammals"]) *
    len(df[(df["Live in Water"]=="yes") & (df["Class"]=="non-mammals")]) / len(df[df["Class"]=="non-mammals"]) *
    len(df[(df["Have Legs"]=="no") & (df["Class"]=="non-mammals")]) / len(df[df["Class"]=="non-mammals"])
)

print("Prediction:", "mammals" if mammals_score > non_score else "non-mammals")
```
