```python
import numpy as np

def kmeans(X, k=3, iters=100):
    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(iters):
        # assign clusters
        labels = [np.argmin([np.linalg.norm(x - c) for c in centroids]) for x in X]

        # update centroids
        new_centroids = []
        for i in range(k):
            points = X[np.array(labels) == i]
            new_centroids.append(points.mean(axis=0) if len(points) > 0 else centroids[i])

        centroids = np.array(new_centroids)

    return np.array(labels), centroids
```

```python
import numpy as np

def kmeans_progress(X, k=3, iters=10):
    centroids = X[np.random.choice(len(X), k, replace=False)]
    history = [centroids.copy()]

    for _ in range(iters):
        labels = np.array([
            np.argmin([np.linalg.norm(x - c) for c in centroids])
            for x in X
        ])

        centroids = np.array([
            X[labels == i].mean(axis=0) if np.any(labels == i) else centroids[i]
            for i in range(k)
        ])

        history.append(centroids.copy())

    return labels, np.array(history)
    
import matplotlib.pyplot as plt

labels, history = kmeans_progress(X, k=3, iters=5)

for i in range(len(history) - 1):
    plt.scatter(history[i][:, 0], history[i][:, 1], c='blue', alpha=0.5)
    plt.scatter(history[i+1][:, 0], history[i+1][:, 1], c='red')

plt.title("KMeans Centroid Movement")
plt.show()
```

```python
labels, history = kmeans_progress(X, k=3, iters=5)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')

for centroids in history:
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x')

plt.title("KMeans Progress")
plt.show()
```

```python
import numpy as np

def kmeans(X, k=3, iters=100):
    np.random.seed(42)
    
    # random initial centroids
    centroids = X[np.random.choice(len(X), k, replace=False)]

    for _ in range(iters):
        # assign clusters
        labels = np.argmin(
            np.linalg.norm(X[:, None] - centroids, axis=2),
            axis=1
        )

        # update centroids
        new_centroids = np.array([
            X[labels == i].mean(axis=0) if len(X[labels == i]) > 0 else centroids[i]
            for i in range(k)
        ])

        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return labels, centroids
```

```python
def silhouette_score_scratch(X, labels):
    n = len(X)
    score = 0

    for i in range(n):
        same_cluster = X[labels == labels[i]]
        other_clusters = X[labels != labels[i]]

        a = np.mean(np.linalg.norm(same_cluster - X[i], axis=1)) if len(same_cluster) > 1 else 0
        b = np.min([
            np.mean(np.linalg.norm(X[labels == j] - X[i], axis=1))
            for j in np.unique(labels) if j != labels[i]
        ])

        score += (b - a) / max(a, b + 1e-9)

    return score / n
```
