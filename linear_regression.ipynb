# Gradient Descent

```python

w = np.zeros(X_b.shape[1])
b = 0
```

```python
w, b
```

```python
losses = []
```

```python
epochs = 100
X = X_b
lr = 0.01
n = len(X)
y = y_train.values.flatten()
for _ in range(epochs):
    z = X @ w + b # x*a1 + a0
    y_hat = z
    error = y_hat - y
    w -= lr * (X.T @ error) / n
    b -= lr * error.mean()
    loss = -np.mean(y*np.log(y_hat+1e-9) + (1-y)*np.log(1-y_hat+1e-9))
    losses.append(loss)
```

```python
class LinearRegressionGD:
    def __init__(self, lr=0.01, epochs=1000):
        self.lr = lr
        self.epochs = epochs

    def fit(self, X, y):
        X = np.array(X)
        y = np.array(y).flatten()

        n, d = X.shape

        # initialize weights and bias separately
        self.w = np.zeros(d)
        self.b = 0.0

        self.losses = []

        for _ in range(self.epochs):
            y_hat = X @ self.w + self.b

            error = y_hat - y

            # gradients
            dw = (X.T @ error) / n
            db = error.mean()

            # update
            self.w -= self.lr * dw
            self.b -= self.lr * db

            # MSE loss
            loss = np.mean(error ** 2)
            self.losses.append(loss)

    def predict(self, X):
        X = np.array(X)
        return X @ self.w + self.b
```
