```python
def edit_distance(word1, word2):
    m, n = len(word1), len(word2)
    
    # Create a matrix of size (m+1) x (n+1)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    
    # Base cases: transforming a string to an empty string
    for i in range(m + 1):
        dp[i][0] = i  # Deletions needed
    for j in range(n + 1):
        dp[0][j] = j  # Insertions needed
        
    # Fill the matrix
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if word1[i-1] == word2[j-1]:
                # Characters match; no new operation needed
                dp[i][j] = dp[i-1][j-1]
            else:
                # Minimum of (Replace, Delete, Insert) + 1
                dp[i][j] = 1 + min(dp[i-1][j-1],  # Replace
                                   dp[i-1][j],    # Delete
                                   dp[i][j-1])    # Insert
                                   
    return dp[m][n], dp

# Example Usage:
ans, m = edit_distance("horse", "ros")  # Output: 3
```
