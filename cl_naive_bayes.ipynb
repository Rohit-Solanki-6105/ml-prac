```python
import numpy as np
import pandas as pd

class NaiveBayes:
    def fit(self, X, y):
        self.X = X.reset_index(drop=True)
        self.y = y.reset_index(drop=True)
        self.classes = self.y.unique()

        # priors
        self.priors = {
            c: len(self.y[self.y == c]) / len(self.y)
            for c in self.classes
        }

        # store subsets
        self.subsets = {
            c: self.X[self.y == c]
            for c in self.classes
        }

    def prob(self, col, val, subset):
        # Laplace smoothing
        count = len(subset[subset[col] == val])
        total = len(subset)
        unique_vals = len(self.X[col].unique())

        return (count + 1) / (total + unique_vals)

    def predict_one(self, x):
        scores = {}

        for c in self.classes:
            score = self.priors[c]
            subset = self.subsets[c]

            for col in self.X.columns:
                score *= self.prob(col, x[col], subset)

            scores[c] = score

        return max(scores, key=scores.get)

    def predict(self, X_test):
        return np.array([self.predict_one(x) for _, x in X_test.iterrows()])
```
